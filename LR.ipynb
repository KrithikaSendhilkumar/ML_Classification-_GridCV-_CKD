```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
# Changing  categorical data to Numerical data
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
dataset
```

```python
# Extracting colunm names
dataset.columns
```

```python
#calculate the no of parameters to be classified in the output
dataset["classification_yes"].value_counts()
```

```python
# Defining the inputs and output
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
independent.shape
```

```python
dependent = dataset[["classification_yes"]]
dependent
```

```python
# Splitting train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
# fitting model for Grid Search
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = {
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'penalty': ['l2']
}

grid = GridSearchCV(
    LogisticRegression(),
    param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1,
    scoring='f1_weighted'
)

grid.fit(x_train,y_train)
```

```python
# print best model after tuning
print(grid.best_params_)
re = grid.cv_results_
```

```python
# print result
grid_predictions = grid.predict(x_test)
```

```python
# Evaluation Metrics
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,grid_predictions)
print(cm)
```

```python
#classification report
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
print(clf_report)
```

```python
#Print f1_score
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test,grid_predictions, average = 'weighted')
print(f1_macro)
```

```python
# Receiver Operating Curve ROC
from sklearn.metrics import roc_auc_score
# ROC AUC Score (using positive class probability)
roc_auc = roc_auc_score(y_test, grid.predict_proba(x_test)[:, 1])
print("ROC AUC Score:", roc_auc)

# Convert GridSearchCV results to DataFrame
table = pd.DataFrame.from_dict(grid.cv_results_)
table.head()
```
